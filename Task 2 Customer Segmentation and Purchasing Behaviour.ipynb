```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by LIFESTAGE and PREMIUM_CUSTOMER to get total spend
total_spend = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=total_spend, x='TOT_SALES', y='LIFESTAGE', hue='PREMIUM_CUSTOMER')
plt.title('Total Spend by Customer Segment and Premium Category')
plt.xlabel('Total Spend')
plt.ylabel('Customer Segment (Lifestage)')
plt.tight_layout()
plt.show()

```

```python
# Average Spend per Transaction by Lifestage and Premium Category
```

```python
# Group and calculate average spend per transaction
avg_spend = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_spend, x='TOT_SALES', y='LIFESTAGE', hue='PREMIUM_CUSTOMER')
plt.title('Average Spend per Transaction')
plt.xlabel('Average Spend')
plt.ylabel('Customer Segment (Lifestage)')
plt.tight_layout()
plt.show()
```

```python
# Total Quantity of Products Purchased by Customer Segment
```

```python
# Group and calculate total quantity purchased
total_qty = data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['PROD_QTY'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=total_qty, x='PROD_QTY', y='LIFESTAGE', hue='PREMIUM_CUSTOMER')
plt.title('Total Product Quantity by Segment')
plt.xlabel('Total Quantity')
plt.ylabel('Customer Segment (Lifestage)')
plt.tight_layout()
plt.show()
```

```python
#Pie Chart â€“ Total Spend Distribution by Lifestage
```

```python
# Total spend by LIFESTAGE
spend_pie = data.groupby('LIFESTAGE')['TOT_SALES'].sum()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(spend_pie, labels=spend_pie.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Spend Distribution by Lifestage')
plt.axis('equal')
plt.show()
```

```python
#Top 10 Brands by Total Sales
```

```python
# Group by brand and get total sales
top_brands = data.groupby('BRAND')['TOT_SALES'].sum().sort_values(ascending=False).head(10).reset_index()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_brands, x='TOT_SALES', y='BRAND', palette='viridis')
plt.title('Top 10 Brands by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Brand')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
transaction = pd.read_csv("QVI_data.csv")

# Create a BRAND_NAME column by extracting first word from PROD_NAME
transaction["BRAND_NAME"] = transaction["PROD_NAME"].str.split().str[0]

# Group by brand and sum the total sales
brand_sales = transaction.groupby("BRAND_NAME")["TOT_SALES"].sum().sort_values(ascending=False)

# Plot the top 10 brands by total sales
plt.figure(figsize=(10, 6))
brand_sales.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Brands by Total Sales')
plt.xlabel('Brand')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
