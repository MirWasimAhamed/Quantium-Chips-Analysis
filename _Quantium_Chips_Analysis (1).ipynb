```python
import pandas as pd
```

```python
customers=pd.read_csv('purchase_behaviour.csv')
```

```python
customers
```

```python
transaction=pd.read_excel('transaction_data.xlsx')
```

```python
transaction
```

```python
print("Missing values in transaction Data:")
print(transaction.isnull().sum())

print("\nMissing values in customers Data:")
print(customers.isnull().sum())
```

```python
print("Duplicate rows in Transaction Data:", transaction.duplicated().sum())
print("Duplicate rows in Customer Data:", customers.duplicated().sum())
```

```python
transactions = transaction.drop_duplicates()
customers = customers.drop_duplicates()
```

```python
transactions
```

```python
print(transaction.dtypes)
print(customers.dtypes)
```

```python
transaction["DATE"] = pd.to_datetime(transaction["DATE"])
```

```python
transaction
```

```python
print(transaction.describe())
```

```python
transaction["PACK_SIZE"] = transaction["PROD_NAME"].str.extract(r"(\d+)\s*G").astype(float)
```

```python
transaction
```

```python
transaction["BRAND_NAME"] = transaction["PROD_NAME"].str.split().str[0]
```

```python
transaction
```

```python
display(transaction.head())
```

```python
import re

# Extract Pack Size
transaction["PACK_SIZE"] = transaction["PROD_NAME"].apply(lambda x: re.search(r"(\d+)[gG]", x))
transaction["PACK_SIZE"] = transaction["PACK_SIZE"].apply(lambda x: float(x.group(1)) if x else None)

# Display results
display(transactions.head())
```

```python
#   Exploratory Data Analysis (EDA)
```

```python
print(transactions.describe())
```

```python
print("Unique Brands in Data:", transaction["BRAND_NAME"].unique())      #Check Unique Brands
```

```python
import matplotlib.pyplot as plt       #Sales Distribution by Pack Size
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(data=transactions, x="PACK_SIZE", order=transactions["PACK_SIZE"].value_counts().index, palette="viridis")
plt.xticks(rotation=90)
plt.title("Pack Size Distribution")
plt.xlabel("Pack Size (grams)")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
brand_sales = transaction.groupby("BRAND_NAME")["TOT_SALES"].sum().sort_values(ascending=False)    #Total Sales by Brand
plt.figure(figsize=(12, 6))
brand_sales.plot(kind="bar", color="orange")
plt.title("Total Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Total Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
merged_df = transaction.merge(customers, on="LYLTY_CARD_NBR", how="left")
print(merged_df.columns)  
```

```python
print(merged_df["LIFESTAGE"].unique())
print(merged_df["PREMIUM_CUSTOMER"].unique())
```

```python
# Data Exploration 
```

```python
sales_by_lifestage = merged_df.groupby("LIFESTAGE")["TOT_SALES"].sum().sort_values(ascending=False)
print(sales_by_lifestage)
```

```python
sales_by_lifestage
```

```python
avg_spend = merged_df.groupby(["LIFESTAGE", "PREMIUM_CUSTOMER"])["TOT_SALES"].mean().unstack()
print(avg_spend)
```

```python
#Visualizing Data & final report
```

```python
import matplotlib.pyplot as plt 
sales_by_lifestage.plot(kind="bar", figsize=(8, 8), color="skyblue")
plt.title("Total Sales by Lifestage")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.show()
```

```python
#Plot Average Spending per Transaction by Lifestage & Premium Category
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(avg_spend, annot=True, cmap="Blues", fmt=".2f")
plt.title("Average Spending per Transaction by Lifestage & Premium Customer")
plt.ylabel("Lifestage")
plt.xlabel("Premium Customer")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("QVI_data.csv")

# Group by brand and sum the sales
brand_sales = data.groupby('brand')['TOT_SALES'].sum().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
brand_sales.head(10).plot(kind='bar', color='skyblue')  # top 10 brands
plt.title('Total Sales by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
